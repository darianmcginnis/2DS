```python
import pandas as pd

value = pd.read_csv('value.csv')
values = value.sort_values('Estimate; Total: - $1,000,000 or more', axis = 0, ascending=False)
print(values)
```

### Top 3 Towns With Highest Amount Houses Worth 1 Million or More
1. Squirrel Hill North
2. Shadyside
3. Mount Washington


```python
values2 = value.sort_values('Estimate; Total: - Less than $10,000', axis = 0, ascending=False)
print(values2)
```

### Top 3 Towns With Highest Amount Houses Worth 10k or Less
1. Perry South
2. Brookline
3. Allentown

### Looking at Crime Values Compared to Housing Values

```python
#Crime Data
crime = pd.read_csv('crime.csv')

#Sorted Alphabetically
crime = crime.sort_values('INCIDENTNEIGHBORHOOD')['INCIDENTNEIGHBORHOOD'].value_counts(sort = False).rename_axis("Neighborhood")
print(crime.to_string())
```

#### HIGHEST HOUSE VALUES COMPARED TO CRIME RATE
* Squirrel Hill North: 39
* Shadyside: 54
* Mount Washington: 72

#### LOWEST HOUSE VALUES COMPARED TO CRIME RATE
* Perry South: 48
* Brookline: 76
* Allentown: 44

### Looking at Population Data Compared to Housing Values
